```python
### Librerias necesarias

#Libreria del sistema
import os.path

#Libreria del URL y request http
import urllib.request

#Libreria del Tiempo
from time import time

# URL del DATASET
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'

# Dirección donde se guardará el archivo y el nombre del archivo
PATH = './datasets/adult.data'
```

```python
# Pregunto si el archivo existe y si tengo acceso o no
if os.path.isfile(PATH) and os.access(PATH, os.R_OK):
    
    # Manda un mensaje para descargar el dataset
    print('El dataset ya está descargado.')
    
else:
    
    # Manda un mensaje informando que lo está descargando
    print('Descargando el dataset...')
    
    # Ejecuto un timer para obtener el tiempo en que se descargó el archivo
    start_time = time()
    
    # Usa la librería urlib para hacer un request y obtener algo de ahí, los parametros son: ( url, nombreArchivo )
    urllib.request.urlretrieve(url, "./datasets/adult.data")
    
    # Calculo el tiempo que tomó en descargar el archivo
    elapsed_time = time() - start_time
    
    #Muestro el tiempo en que se descargó
    print("Se ha descargado el dataset en:", elapsed_time, "segundos.")
```

```python
# Creo las columnas del dataset, ya que por defecto no las tiene
columnas = ['age','workclass','fnlwgt','education','education-num','marital-status','ocupation',
            'relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','target']
```

```python
import pandas as pd
import numpy as np
```

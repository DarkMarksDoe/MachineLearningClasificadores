```python


### Librerias necesarias

#Libreria del sistema
import os.path

#Libreria del URL y request http
import urllib.request

#Libreria del Tiempo
from time import time

# URL del DATASET
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'

# Dirección donde se guardará el archivo y el nombre del archivo
PATH = './datasets/adult.data'
```

```python
# Pregunto si el archivo existe y si tengo acceso o no
if os.path.isfile(PATH) and os.access(PATH, os.R_OK):
    
    # Manda un mensaje para descargar el dataset
    print('El dataset ya está descargado.')
    
else:
    
    # Manda un mensaje informando que lo está descargando
    print('Descargando el dataset...')
    
    # Ejecuto un timer para obtener el tiempo en que se descargó el archivo
    start_time = time()
    
    # Usa la librería urlib para hacer un request y obtener algo de ahí, los parametros son: ( url, nombreArchivo )
    urllib.request.urlretrieve(url, './datasets/adult.data')
    
    # Calculo el tiempo que tomó en descargar el archivo
    elapsed_time = time() - start_time
    
    #Muestro el tiempo en que se descargó
    print('Se ha descargado el dataset en:', elapsed_time, 'segundos.')
```

```python
# Creo las columnas del dataset, ya que por defecto no las tiene
columnas = ['age','workclass','fnlwgt','education','education-num','marital-status','ocupation',
            'relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','target']
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(PATH, sep=', ',header=None)
```

```python
df.head(50)
```

```python
df.columns = columnas
```

```python
df.head(50)
```

```python
df['target'].unique()
```

```python
df['target'] = np.where(df['target'] == '<=50K',0,1)
```

```python
df['target'].unique()
```

```python
df.head(50)
```

```python
df.shape
```

```python
#Ingenieria de Variable
df['hours-per-week'].value_counts()
```

```python
df['m-i-40'] = np.where (df['hours-per-week'] <= 40, 1, 0)
df.head()
```

```python
df['dif-capital'] = df['capital-gain'] -  df['capital-loss']
df.head()
```

```python
df['sex'] = np.where(df['sex'] == 'Female', 1, 0)
df.head()
```

```python
df['mexico'] = np.where(df['native-country'] == 'Mexico', 1, 0)
df.head(50)
```

```python
df['workclass'].value_counts()
```

```python
a = df.groupby('workclass')['workclass'].transform('count')
a
```

```python
cat_col = ['workclass', 'education', 'marital-status', 'ocupation', 'relationship', 'race', 'native-country']
```

```python
for cat in cat_col:
    df[cat] = df.groupby(cat)[cat].transform('count')
```

```python
df.head(50)
```

```python
X = df.loc[:, df.columns != 'target']
```

```python
y = df.loc[: , 'target']
```

```python
X.shape, y.shape, df.shape
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size = 0.5, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
X_valid, X_test, y_valid, y_test = train_test_split(X_test, y_test, test_size = 0.5, random_state=1)
```

```python
X_train.shape, X_test.shape, X_valid.shape
```

```python
!pip install xgboost
import xgboost
```

```python
xgb = xgboost.XGBClassifier()
```

```python
parameters = {
    'nthreads' : [3],
    'objective': ['binary:logistic'],
    'learning_rate' : [0.3],
    'm_estimators': [100,200,500],
    'max_depth': 4,
    'base_score':0.5
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
fit_params = {
    'verbose': [False],
    'early_stopping_rounds' : 20,
    'eval_metric' : 'logloss',
    'eval_set': [(X_test, y_test)]
}
```

```python
grid_params ={  
    'max_depth': [1, 2, 3, 4, 5],
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],
    'subsample': [0.7, 0.8, 0.9, 1.0],    
}
```

```python
clf = GridSearchCV(
    xgb, 
    cv=5,
    param_grid  = grid_params,
    scoring = 'accuracy'
)
```

```python
clf.fit(
    X_train,
    y_train,
    **fit_params
)
```

```python
clf.best_estimator_
```

```python
clf.best_score_
```

```python
from sklearn.metrics import accuracy_score
```

```python
best_xgb = clf.best_estimator_
```

```python
y_preds = best_xgb.predict(X_valid)
```

```python
comp = pd.DataFrame({'real':y_valid, 'preds' : y_preds})
```

```python
comp.head(20)
```

```python
acc = accuracy_score(y_valid, y_preds)
```
